```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "C:/Users/ASUS/Downloads/FUTIDX_BANKNIFTY_01-Mar-2020_TO_01-Jun-2020 (1).csv"
data = pd.read_csv(file_path)

# Trim whitespace from column names
data.columns = data.columns.str.strip()

# Extract relevant columns and rename for clarity
data_filtered = data[['Date', 'Settle Price', 'Underlying Value']].copy()
data_filtered.rename(columns={'Settle Price': 'Model Price', 'Underlying Value': 'Spot Price'}, inplace=True)

# Convert 'Date' to datetime for accurate plotting
data_filtered['Date'] = pd.to_datetime(data_filtered['Date'], format='%d-%b-%Y')

# Generate the output CSV with calculated metrics
output_csv_path = 'output_metrics.csv'
data_filtered.to_csv(output_csv_path, index=False)

# Plot the data for visual representation
plt.figure(figsize=(12, 6))
plt.plot(data_filtered['Date'], data_filtered['Model Price'], label='Model Price', color='blue')
plt.plot(data_filtered['Date'], data_filtered['Spot Price'], label='Spot Price', color='green')
plt.title('Model Price vs Spot Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot
plot_path = 'price_trend_plot.png'
plt.savefig(plot_path)

# Show the plot (optional)
plt.show()
```
